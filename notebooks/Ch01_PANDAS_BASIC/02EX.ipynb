```python
import pandas as pd

df = pd.DataFrame({
    "메뉴": ['아메리카노', '카페라떼', '카페모카', '카푸치노', '에스프레소', '밀크티', '녹차'],
    "가격": ["4500", "NA", "5500", "5000", "-", "5900", "-"],
    "칼로리": [10, 110, 250, 110, 20, 210, 0],
})
df
```

```python
#DataFrame df를 cafe_basic.csv로 저장하고, 이를 다시 읽어 df_basic에 담아(인덱스 포함)
import pandas as pd
df.to_csv("cafe_basic.csv")
df_basic = pd.read_csv
```

```python
# df를 인덱스 없이 cafe_no_index.csv로 저장하고, 읽어온 DataFrame을 df_no_index로.
df.to_csv("cafe_no_index",index = False)
df1=pd.read_csv("cafe_no_index")
df1
```

```python
#메뉴, 가격 컬럼만 cafe_menu_price.csv로 저장하시오. 인덱스는 제외.
df.to_csv("cafe_menu_price",index = False,columns=['메뉴','가격'])
df2=pd.read_csv("cafe_menu_price")
df2
```

```python
# 세미콜론(;) 구분자로 cafe_sc.csv를 저장하고, 같은 구분자로 읽어 df_sc에 담으시오
df.to_csv("cafe_sc.csv", sep=";",index=False)
df_sc=pd.read_csv("cafe_sc.csv",sep=";")
df_sc
```

```python
# Windows Excel에서 한글이 깨지지 않도록 utf-8-sig 인코딩으로 cafe_utf8.csv를 저장하시오.
df.to_csv("cafe_utf8.csv", index=False, encoding="utf-8-sig")
df3=pd.read_csv("cafe_utf8.csv")
df3
```

```python
# 헤더 없이 cafe_noheader.csv로 저장한 뒤, header=None으로 읽어 df_noheader로 저장
df.to_csv("cafe_noheader.csv",index=False,header=False)
df_noheader = pd.read_csv("cafe_noheader.csv", header=None)
df_noheader
```
